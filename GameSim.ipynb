```python
import pandas as pd
import numpy as np

nba_game_sim = pd.read_csv('c:\\Users\\moholom\\Desktop\\NBA\\Excel\\Projections\\NBA Projections - Sim.csv')
final_score = []
win_count = []
spread = []

home = nba_game_sim[nba_game_sim['Team'] == "Home"]
away = nba_game_sim[nba_game_sim['Team'] == 'Away']

for i in range(10000):

    home_total_ppm = []
    for p in range(len(home.index)):
        h_player_point_ppm = (np.random.normal(home.iloc[p]['PTS/MIN'], home.iloc[p]['STDEV']))
        if h_player_point_ppm < 0:
            h_player_point_ppm = 0
        home_total_ppm.append(h_player_point_ppm)

    h_rem_mins = 240
    h_possible_max = 48
    h_game_score = []
    np.random.shuffle(home_total_ppm)
    for i in home_total_ppm:

        h_relevant = np.min([h_rem_mins, h_possible_max])
    
        if h_relevant == 0:
            h_player_min = 0
        else:
            if i == home_total_ppm[-1]:
                h_player_min = h_relevant
    
            else:
                h_player_min = np.random.normal(np.median(range(h_relevant))*1.25, 10)

            if h_player_min > h_relevant:
                h_player_min = h_relevant
            elif h_player_min < 0:
                h_player_min = 0


        h_game_score.append((((home['HCA'].mean() * ((int(h_player_min)) / 240))**(5.85) + i) * int(h_player_min)))
        h_rem_mins -= int(h_player_min)


    away_total_ppm = []
    for q in range(len(away.index)):
        a_player_point_ppm = (np.random.normal(away.iloc[q]['PTS/MIN'], away.iloc[q]['STDEV']))
        if a_player_point_ppm < 0:
            a_player_point_ppm = 0
        away_total_ppm.append(a_player_point_ppm)

    a_rem_mins = 240
    a_possible_max = 48
    a_game_score = []
    np.random.shuffle(away_total_ppm)
    for i in away_total_ppm:

        a_relevant = np.min([a_rem_mins, a_possible_max])
    
        if a_relevant == 0:
            a_player_min = 0
        else:
            if i == away_total_ppm[-1]:
                a_player_min = a_relevant
    
            else:
                a_player_min = np.random.normal(np.median(range(a_relevant))*1.25, 10)

            if a_player_min > a_relevant:
                a_player_min = a_relevant
            elif a_player_min < 0:
                a_player_min = 0


        a_game_score.append(i * int(a_player_min))
        a_rem_mins -= int(a_player_min)

    final_score.append((sum(h_game_score)) + sum(a_game_score))
    if (sum(h_game_score)) > sum(a_game_score):
        win_count.append(1)
    else:
        win_count.append(-1)
    spread.append((sum(h_game_score)) - sum(a_game_score))
    

print("Home Win % - ", (win_count.count(1))/len(win_count))
print("Away Win % - ", (win_count.count(-1))/len(win_count))
print("Average Final Score - ", (np.mean(final_score)))
print("Average Home Margin of Victory - ", ((np.mean(spread))/2.5))
```
