```python
import pandas as pd
import numpy as np

nba_szn_sked = pd.read_csv('c:\\Users\\moholom\\Desktop\\NBA\\Excel\\nba_schedule.csv')
nba_team_infors = pd.read_csv('c:\\Users\\moholom\\Desktop\\NBA\\Excel\\NBA_Team_Infors.csv')


wins = pd.DataFrame(columns=list(nba_team_infors['Team'].unique()))
for j in range(10000):

    season_wins = []
    szn = pd.DataFrame()
    for k in range(len(nba_szn_sked)):
        home = nba_team_infors[nba_team_infors['Team'] == nba_szn_sked.iloc[k]['Home']]
        away = nba_team_infors[nba_team_infors['Team'] == nba_szn_sked.iloc[k]['Away']]

        game = nba_szn_sked.iloc[[k]].copy()

        home_total_ppm = []
        for p in range(len(home.index)):
            h_player_point_ppm = (np.random.normal(home.iloc[p]['PTS/MIN'], home.iloc[p]['STDEV']))
            if h_player_point_ppm < 0:
                h_player_point_ppm = 0
            home_total_ppm.append(h_player_point_ppm)

        h_rem_mins = 240
        h_possible_max = 48
        h_game_score = []
        np.random.shuffle(home_total_ppm)
        for h in home_total_ppm:

            h_relevant = np.min([h_rem_mins, h_possible_max])
    
            if h_relevant == 0:
                h_player_min = 0
            else:
                if h == home_total_ppm[-1]:
                    h_player_min = h_relevant
    
                else:
                    h_player_min = np.random.normal(np.median(range(h_relevant))*1.25, 10)

                if h_player_min > h_relevant:
                    h_player_min = h_relevant
                elif h_player_min < 0:
                    h_player_min = 0


            h_game_score.append(h * int(h_player_min))
            h_rem_mins -= int(h_player_min)

        away_total_ppm = []
        for q in range(len(away.index)):
            a_player_point_ppm = (np.random.normal(away.iloc[q]['PTS/MIN'], away.iloc[q]['STDEV']))
            if a_player_point_ppm < 0:
                a_player_point_ppm = 0
            away_total_ppm.append(a_player_point_ppm)

        a_rem_mins = 240
        a_possible_max = 48
        a_game_score = []
        np.random.shuffle(away_total_ppm)
        for a in away_total_ppm:

            a_relevant = np.min([a_rem_mins, a_possible_max])
    
            if a_relevant == 0:
                a_player_min = 0
            else:
                if a == away_total_ppm[-1]:
                    a_player_min = a_relevant
    
                else:
                    a_player_min = np.random.normal(np.median(range(a_relevant))*1.25, 10)

                if a_player_min > a_relevant:
                    a_player_min = a_relevant
                elif a_player_min < 0:
                    a_player_min = 0


            a_game_score.append(a * int(a_player_min))
            a_rem_mins -= int(a_player_min)

        game['Points - H'] = sum(h_game_score)
        game['Points - A'] = sum(a_game_score)
        game['HCA'] = home.iloc[0]['HCA']
        
        if sum(h_game_score) + game['HCA'].mean() > sum(a_game_score):
            game['Winner'] = game.iloc[0]['Home']
        else:
            game['Winner'] = game.iloc[0]['Away']

        frames = [szn,game]
        szn = pd.concat(frames)

    
    for team in wins.columns:
        season_wins.append((szn.Winner.value_counts()[team]/80)*82)
    season_wins = pd.DataFrame(season_wins).transpose()
    season_wins.columns = list(nba_team_infors['Team'].unique())
    wins = pd.concat([wins,season_wins])

average_wins = []
for team in wins.columns:
    average_wins.append(wins[team].mean())
average_wins = pd.DataFrame(average_wins).transpose()
average_wins.columns = list(nba_team_infors['Team'].unique())
average_wins.to_csv('c:\\Users\\moholom\\Desktop\\NBA\\Excel\\NBA_RSW.csv')
```
